```python
df.info()
```

#### Visulaizacion del target

```python
base = df[['FLAG_VENTA']]
```

#### Identificar variables categoricas

```python
## identify categorical variables
cat = df.select_dtypes(exclude=['int', 'float']).columns
print(len(cat))
print('Number of categoriacal features:', len(cat))

CatVar=df[cat]
Nl=[]
for column in CatVar.columns:
    if CatVar[column].isnull().values.any() == True:
        Nl.append(column)
    
print('columns with missing values:', Nl)
```

```python
for column in CatVar.columns:
    print(CatVar.groupby(column)[column].agg(['count']))
```

```python
CatVar
```

```python
CatVar['SEXO']=np.where(CatVar['SEXO']=='M', 'masculino',
                            np.where(CatVar['SEXO']=='F', 'femenino',
                                   CatVar['SEXO']))

CatVar['EST_CIVIL']=np.where(CatVar['EST_CIVIL']=='C', 'casado',
                         np.where(CatVar['EST_CIVIL']=='S', 'soltero',
                                np.where(CatVar['EST_CIVIL']=='D', 'divorciado',
                                        np.where(CatVar['EST_CIVIL']=='U', 'otros',
                                             np.where(CatVar['EST_CIVIL']=='V', 'otros',
                                                np.where(CatVar['EST_CIVIL']=='X', 'otros',        
                                                  np.where(CatVar['EST_CIVIL']=='Y', 'otros',                                                         
                                                    CatVar['EST_CIVIL'])))))))
```

```python
# Revisar los valores perdidos
#Creamos una funciÃ³n:
def num_missing(x):
  return sum(x.isnull())

#Aplicamos por columna:
print ("Valores perdidos por columna")
print (CatVar.apply(num_missing, axis=0))
```

```python
## imputacion por moda
CatVar = CatVar.fillna(CatVar.mode().iloc[0])
```

```python
for column in CatVar.columns:
    print(CatVar.groupby(column)[column].agg(['count']))
```

```python
Cat_dummies = pd.get_dummies(CatVar, drop_first=True)
print(CatVar.shape)
print(Cat_dummies.shape)
```

#### Variables cuantitativas

```python
## Identify continuous variables
cont1 = df.select_dtypes(include=['int', 'float']).columns
print('initial number of continuous features: ',len(cont1))


## Exclude those previously classified as dummies
cont4=set(cont1).difference(cat)
cont=set(cont4).difference(base)
print('initial number of continuous features after exclusions: ',len(cont))
print()

ContVar= df[cont]

Nl3=[]
for column in ContVar.columns:
    if ContVar[column].isnull().values.any() == True:
        Nl3.append(column)
    
print('features with missing values: ',Nl3)
```

```python
# Revisar los valores perdidos
#Creamos una funciÃ³n:
def num_missing(x):
  return sum(x.isnull())

#Aplicamos por columna:
print ("Valores perdidos por columna")
print (ContVar.apply(num_missing, axis=0))
```

```python
ContVar['SUELDO_ESTIMADO'].fillna(ContVar['SUELDO_ESTIMADO'].median(), inplace = True)
ContVar['USO_LINEA_TOTAL_TC_T2'].fillna(ContVar['USO_LINEA_TOTAL_TC_T2'].median(), inplace = True)
ContVar['ANTIGUEDAD_MES'].fillna(ContVar['ANTIGUEDAD_MES'].median(), inplace = True)
ContVar['EDAD'].fillna(ContVar['EDAD'].median(), inplace = True)
ContVar['RESENCIA_OFERTA_PLD_RENOVADO'].fillna(ContVar['RESENCIA_OFERTA_PLD_RENOVADO'].mean(), inplace = True)
ContVar['SDO_CONSUMO_T2'].fillna(ContVar['SDO_CONSUMO_T2'].mean(), inplace = True)
ContVar['USO_TRIM_LINEA_BBVA'].fillna(ContVar['USO_TRIM_LINEA_BBVA'].median(), inplace = True)
```

```python
df2 = pd.concat([base,ContVar,Cat_dummies],axis=1)
df2.shape
df2.to_csv('data_clean.csv', index=False) # make sure you drop index when saving to csv
print(df2.shape)
```
